### Scatter plot

```python
# Load data into a pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
# Using a scatter plot
sns.scatterplot(data=flowers_df, x="sepal_length", y="sepal_width");
```

```python
# Adding hues to scatterplot
sns.scatterplot(data=flowers_df,
                x="sepal_length", y="sepal_width",
                hue=flowers_df.species, s=60);
```

## Histogram

```python
plt.title('Distribution of sepal width')
plt.hist(flowers_df.sepal_width);
```

```python
# Specifying bins
plt.title('Distribution of petal width')
plt.hist(flowers_df.petal_width, bins=6);
```

```python
import numpy as np

#Specifying the boundaries of each bin
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25));
plt.legend(['Setosa','Versicolor'])
```

```python
# We can stack multiple histograms on top of one another

plt.title('Distribution of sepal width')

plt.hist([setosa_df.sepal_width,virginica_df.sepal_width,versicolor_df.sepal_width],
         bins=np.arange(2,5,0.25),
         stacked=True);
plt.legend(['Setosa','Versicolor','Virginica'])
```

```python
import plotly.express as px
fig = px.scatter_3d(flowers_df, x='sepal_length', y='sepal_width', z='petal_length', color='species')
fig.update_layout(title='3D Scatter Plot for Iris Dataset')
fig.show()
```

```python
# Calculate the frequency of each species
species_counts = flowers_df['species'].value_counts()
```

```python
# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', shadow=True, startangle=45)
plt.title('Distribution of Iris Species')
plt.axis('equal')
# Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Create a pair plot
sns.pairplot(flowers_df,hue='species',palette='husl',markers=['o','s','D'])
plt.title('Pair plot for iris dataset')
plt.show()
```

```python
# Creating stem and leaf plots
# Extract the sepal length values
sepal_length_values = setosa_df['sepal_length'].values

# Creating a stem and leaf plot
fig, ax = plt.subplots()
ax.stem(sepal_length_values,
        use_line_collection=True)
ax.set_title('Stem and leaf plot for sepal length(Setosa)')
ax.set_xlabel('Stems')
ax.set_ylabel('Leaves')
plt.show()
```

```python
# Creating a contour plot
sns.kdeplot(x="sepal_length", y="petal_length", data=flowers_df, cmap='viridis',fill=True)
plt.xlabel('Sepal length (cm)')
plt.xlabel('Petal length (cm)')
plt.title('Contour Plot for Iris Dataset')
plt.show()
```

```python
# Creating a contour plot
sns.kdeplot(x="sepal_width", y="petal_width", data=flowers_df, cmap='viridis',fill=True)
plt.xlabel('Sepal width (cm)')
plt.ylabel('Petal length (cm)')
plt.title('Contour Plot for Iris Dataset')
plt.show()
```
